```python
#importing libraries 
import requests
from bs4 import BeautifulSoup

import pandas as pd
import numpy as np
import math

# Scraping data from Tranfermarkt.de - Reference: https://fcpython.com/web-scraping
# Data: 1. Bundesliga, Saison 2019/2020 
headers = {'User-Agent': 
           'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.106 Safari/537.36'}

page = "https://www.transfermarkt.de/1-bundesliga/tabelle/wettbewerb/L1/saison_id/2019"
pageTree = requests.get(page, headers=headers)
pageSoup = BeautifulSoup(pageTree.content, 'html.parser')

# Exctracting teams' and match results  
Teams = pageSoup.find_all("a", {"class": "vereinprofil_tooltip"})

Matches = pageSoup.find_all("td", {"class": "zentriert"})

# Interating over dataset and removing irrelevant data 
TeamsList = []
MatchesList = []
ScoreList = []

for i in range(0,37):
    TeamsList.append(Teams[i].text)
for i in range(0,145):
    MatchesList.append(Matches[i].text)
for i in range(0,146):
    ScoreList.append(Matches[i].text) 
    
del TeamsList[0::2]
del MatchesList[0::2]
del ScoreList[0::8]

MatchesList = MatchesList[0::4]
ScoreList = ScoreList[1::7]

# Creating dataframe for Teams, Wins and number of matches in the season
      
df = pd.DataFrame({"Teams":TeamsList,"Spiele":MatchesList, "Siege":ScoreList}, columns=['Teams', 'Spiele', 'Siege'])
df.head()


df['Siege'] = pd.to_numeric(df['Siege'])
df['Spiele'] = pd.to_numeric(df['Spiele'])

df['win_rate_of_competitors']=df['Siege']/df['Spiele']
print(df)

# var ( real ) = var ( skill ) + var ( luck )
# var ( skill ) = var ( real ) - var ( luck )
# var ( real )  = var ( win_rate_of_competitors )
# var ( luck ) = var ( game_is_a_coin_toss ) 
# where var := standard deviation

# Calculating game is a toss variable (var (luck))
game_is_a_coin_toss = math.sqrt(0.5*(0.5/153))

print("Game is a coin toss:", game_is_a_coin_toss)

#var (real) = 0.167462 = 0.12703939582727783 + 0.040422604172722164
var_skill = 0.167462 - 0.040422604172722164
print("Var(skill): ", var_skill)

#var ( real )  = 0.167462
#var ( luck ) = 0.040422604172722164

luck =  0.040422604172722164 / 0.167462
skill = 0.12703939582727783 / 0.167462

print("Luck:",luck,"Skill:", skill)

print("Standard deviation")
df.std()
```
