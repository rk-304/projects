```r
set.seed(1)
Et <- ts(rnorm(101, 0, 1))
Et[1] <- 0
y1 <- 0 #delete later
for (i in 2:length(Et)) {
  y1[i] <- Et[i]}
y1 = y1[2:length(y1)]
ts.y1 = ts(y1)
plot(ts.y1, main="Xt=Et")
```

```r
# Excersize 3.a) 
#Moving Average - MA(1) Model
set.seed(1)
Et <- ts(rnorm(101, 0, 1))
Et[1] <- 0
y1 <- 0 #delete later
for (i in 2:length(Et)) {
y1[i] <- Et[i]-0.5*Et[i-1]
}
y1 = y1[2:length(y1)]
ts.y1 = ts(y1)
ts.plot(ts.y1, main = 'Exercise 3.a', xlab = 'Time', ylab = '', col='blue', lwd = 2)
#Stationary 
```

```r
# Excersize 3.b) 
#Autoregressive (AR Model) 

set.seed(1) 
y=NULL
y[1]=0
for (i in 2:100) {
   y[i] = y[i-1] + rnorm(1,0,1)
}
ts.plot(y, main = 'Exercise 3.b', xlab = 'Time', ylab = '', col='blue', lwd = 2)
#Non-stationary
```

```r
# Excersize 3.c)
#Autoregressive (AR Model) 

set.seed(1) 
y=NULL
y[1]=0
for (i in 2:100) {
   y[i] = 0.5*y[i-1] + rnorm(1,0,1)
}
ts.plot(y, main = 'Exercise 3.c', xlab = 'Time', ylab = '', col='blue', lwd = 2)
#Stationary
```

```r
# Excersize 3.d)
#Autoregressive (AR Model) 

set.seed(1) 
y=NULL
y[1]=1
for (i in 2:100) {
   y[i] = y[i-1] * rnorm(1,0.95,1.05)
}
ts.plot(y, main = 'Exercise 3.d', xlab = 'Time', ylab = '', col='blue', lwd = 2)
#Non-Stationary
```
